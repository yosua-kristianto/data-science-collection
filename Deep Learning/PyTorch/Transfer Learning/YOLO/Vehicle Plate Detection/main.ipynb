### Configure ROOT_DIR and CONFIG_FILE

> These configuration variables below are used to configure the root directory of the yaml YOLO training configuration file. I have set marked variables below with comments so if you want to change the training configuration file path, you may not want to run the uncorrespondent script.

```python
# This training data setup came from /our-own-training-set folder

ROOT_DIR = 'path\\to\\bda-team-yolo-code\\our-own-training-set'; # Path to our-own-training-set folder
CONFIG_FILE = "config.yaml"; # The config yaml file name
EPOCH = 200; # Training epoch
DATASET_TOTAL = "1204";
```

### Support Functions Part
This function below helps reporting process through Telegram whenever the training process starts, ended up forcefully, or finished.

```python
import requests;
import time;

def telegram_reporter(message):
    message = "["+ time.strftime("%Y-%m-%d %H:%M:%S") +"] " + message
    requests.request(method="POST", url="https://api.telegram.org/bot:botid/sendMessage?chat_id=:chatid&text=" + message, headers={}, data={});
```

### Training Model
In short, yes just train your model.

```python
import os
from ultralytics import YOLO
# Load a model
model = YOLO('yolov8n.yaml')  # build a new model from scratch
model = YOLO('yolov8n.pt')  # load a pretrained model (recommended for training)

report_message = "Training session not yet started...";

# Use the model
try:
    telegram_reporter("Training started with configuration of:\n " 
                      + "ROOT_DIR: " + ROOT_DIR + "\n" 
                      + "CONFIG_FILE: " + CONFIG_FILE + "\n" 
                      + "EPOCH: " + str(EPOCH) + "\n" 
                      + "TOTAL_DATASET: " + DATASET_TOTAL + "\n" 
                      + "Further Description: This training session will stack the previous trained model earlier (20231019.0.0.2)." + "\n" 
                      );
    results = model.train(data=os.path.join(ROOT_DIR, CONFIG_FILE), epochs=EPOCH)  # train the model
    report_message = "Training session finished";
except Exception as e:
    report_message = "Training session has been terminated. Please check your ipynb for further information."
    print(e)

telegram_reporter(report_message);

```
